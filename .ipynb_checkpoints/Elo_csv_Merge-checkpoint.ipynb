```python
# Create a map from card_id to unique int
card_id_map = dict(zip(
    train['card_id'].values,
    train['card_id'].astype('category').cat.codes.values
))

# Map the values
train['card_id'] = train['card_id'].map(card_id_map).astype('uint32')
```
